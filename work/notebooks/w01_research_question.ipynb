## 2. The question: decision, action, cost of a wrong call

*What decision does your work improve? Who acts on it? What does a wrong recommendation cost?*

```python
The real goal is not to predict which pages will decline. The goal is to help the SEO team decide which pages deserve attention first.

The SEO and content teams will use the recommendations to prioritize their work instead of reviewing every page manually.

A wrong recommendation has two costs: reviewing the wrong page wastes valuable time, while missing an important page can allow performance to continue declining.

If a simple rule could reliably make this decision, there would be no reason to build a machine learning model. The value of ML is combining multiple signals to support better decisions, not replacing human judgment.
```
